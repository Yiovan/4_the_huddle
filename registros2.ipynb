```python
import requests
from bs4 import BeautifulSoup

url = 'https://books.toscrape.com/'
respuesta = requests.get(url)
soup = BeautifulSoup(respuesta.content, 'html.parser')

# cada categoría tiene su propio enlace
categorias = soup.find("ul", class_="nav nav-list").find("ul").find_all("li")

for categoria in categorias:
    nombre = categoria.find("a").text.strip()
    link = "https://books.toscrape.com/" + categoria.find("a")["href"]

    respuesta_cat = requests.get(link)
    soup_cat = BeautifulSoup(respuesta_cat.content, 'html.parser')

    libros = soup_cat.find_all('article', class_='product_pod')

    print(f'\n📂 {nombre}')
    for libro in libros:
        titulo = libro.h3.a['title']
        precio = libro.find('p', class_='price_color').text
        print(f'   📚 {titulo} — {precio}')
```
