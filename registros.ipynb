```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://quotes.toscrape.com/"
```

```python
respuesta = requests.get(url)

if respuesta.status_code == 200:
    print('pagina descargada completamente')
else:
    print("error al descargar la pagina")
```

```python
sopa = BeautifulSoup(respuesta.text, "html.parser")
```

```python
frases = sopa.find_all("span", class_="text")

for i, frase in enumerate(frases, start =1):
    print(f'{i} : {frase.text}')
    
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com"
response = requests.get(url)
response.encoding = "utf-8"
soup = BeautifulSoup(response.text, "html.parser")

# Buscar todos los títulos de libros
libros = soup.find_all("article", class_="product_pod")

for libro in libros:
    titulo = libro.find("h3").find("a")["title"]
    precio = libro.find("p", class_="price_color").text
    print(f"📚 {titulo} — {precio}")
```
